```python
import pandas as pd

cheminDuFichierCSV = "./../data/dictionary/1757_Buchanan-J.csv"
buch = pd.read_csv(cheminDuFichierCSV, delimiter=";", encoding="utf8")

cheminDuFichierCSV = "./../data/dictionary/Buchanan_PronChar_counts.csv"
pronchar = pd.read_csv(cheminDuFichierCSV, delimiter=";", encoding="utf8")
```

```python
def motVedette(mot):
    TruePron = []
    transfo = ''
    for lettre in mot:
        if (pronchar[pronchar['char']== lettre]['graph'].isnull().values[0]):
            transfo = ''
        else: transfo = pronchar[pronchar['char']== lettre]['graph'].values[0]
        TruePron.append(transfo)  
    return ''.join(TruePron)
        
def sizePron(strPron):
    return len(strPron)
```

```python
buch['motVedette'] = buch.apply(lambda x: motVedette(x['pron']),1)
buch['sizePron'] = buch.apply(lambda x: sizePron(x['pron']),1)
buch
```

```python
cheminDuFichierCSV = "./../data/dataframe/BigiRichSample.csv"
bigi = pd.read_csv(cheminDuFichierCSV, delimiter=";", encoding="utf8")
```

```python
dfDicoBuch_augmente = buch.merge(bigi, left_on='motVedette', right_on='hwd')
```

```python
path = "./../data/dataframe/buchan.csv"
dfDicoBuch_augmente.to_csv(path, sep=";",encoding="utf8",index=False)
```
